```python
#Imports
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Reading the dataset
df = pd.read_csv('./data/housepricedata.csv')
```

```python
df
```

```python
# Converting the dataframe into an array.
dataset = df.values
```

```python
dataset
```

```python
# Extracting X 
X = dataset[:,0:10]
```

```python
# Y dataset
Y = dataset[:,10]
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
# Array after scaling
X_scale
```

```python
# Using sklearn test_train_split to split data into test and train dataset
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
```

```python
# Usinf train_test_split to create validation data
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
# Printing the shape of all the variables
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
# Define the architecture of the model
model = Sequential([    Dense(32, activation='relu', input_shape=(10,)),    Dense(32, activation='relu'),    Dense(1, activation='sigmoid'),])
```

```python
# Compiling the model
model.compile(optimizer='sgd',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Training your model
hist = model.fit(X_train, Y_train,batch_size=32, epochs=100,validation_data=(X_val, Y_val))
```

```python
# Evaluate the model
model.evaluate(X_test,Y_test)[1]
```
